```python
bins = [-10000, 0, 10000, 35000, 70000,]
labels = ["negative", "low", "mid", "high"]
df["income"] = pd.cut(df["balance"], bins=bins, labels=labels, right=False)
display(df.head())
```

```python
df["income"].value_counts().plot(kind="bar")
```

```python
sns.scatterplot(x='age', y='balance', data=df)
plt.pyplot.title('Scatter Plot of Age vs Balance')
plt.pyplot.xlabel('Age')
plt.pyplot.ylabel('Balance')
plt.pyplot.show()
```

BIVARITE ANALYSIS

```python
categorical_cols = df[["age","job", "marital", "education", "default", "balance", "housing", "loan", "contact", "day", "month", "duration", "campaign", "pdays", "previous", "poutcome", "Target", "income"]].columns

for col in categorical_cols:
  print(f"value count for '{col}':")
  display(df[col].value_counts().plot())
```

BAR CHART

```python
for i, predictor in enumerate(df):
  plt.pyplot.figure(i)
  sns.countplot(data=df, x=predictor, hue="age")
```

```python
for i, predictor in enumerate(df):
  plt.pyplot.figure(i)
  sns.countplot(data=df, x=predictor)
```

```python
plt.pyplot.figure(figsize=(10, 6))
sns.countplot(y='job', data=df, order=df['job'].value_counts().index)
plt.pyplot.title('Distribution of Job Types')
plt.pyplot.xlabel('Count')
plt.pyplot.ylabel('Job Type')
plt.pyplot.show()
```

NUMERICAL ANALISIS

```python
numerical_cols = df.select_dtypes(include=np.number).columns

for col in numerical_cols:
  plt.pyplot.figure(figsize=(8, 6))
  sns.histplot(data=df, x=col)
  plt.pyplot.title(f'Distribution of {col}')
  plt.pyplot.xlabel(col)
  plt.pyplot.ylabel('Frequency')
  plt.pyplot.show()
```

```python
numerical_cols = ["age","job", "marital", "education", "default", "balance", "housing", "loan", "contact", "day", "month", "duration", "campaign", "pdays", "previous", "poutcome", "Target", "income"]
plt.pyplot.figure(figsize=(30, 35))
for i, col in enumerate(numerical_cols):
  plt.pyplot.subplot(6, 3, i+1)
  sns.histplot(df[col], kde=True)
  plt.pyplot.title(col)
plt.pyplot.show()
```

HEATMAP

```python
numerical_cols = ["age", "balance", "day", "duration", "campaign", "pdays", "previous"]

correlation_matrix = df[numerical_cols].corr()
plt.pyplot.figure(figsize=(15, 15))
sns.heatmap(correlation_matrix,annot=True, cmap="crest", fmt=".2f")
plt.pyplot.title("correlation matrix")
plt.pyplot.show()
```
